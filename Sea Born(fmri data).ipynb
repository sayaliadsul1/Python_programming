```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
fmri=sns.load_dataset("fmri")
fmri
```

#  Line plot

```python
sns.lineplot( x = "timepoint",y = "signal",data = fmri)
```

```python
sns.lineplot( x = "timepoint",y = "signal",hue = "region",data = fmri)
```

```python
sns.lineplot( x = "timepoint",y = "signal",hue = "region",style = "event",data = fmri)
```

```python
sns.lineplot( x = "timepoint",y = "signal",size = "event",hue = "region",data = fmri)
```

```python
sns.lineplot(x = "timepoint", y = "signal",hue = "subject",style = "event",data = fmri)
```

```python
# adding palette, style and hue semantic
sns.lineplot( x = "timepoint", y = "signal",
             hue = "subject", style = "event",
             palette = "YlOrRd_r", data = fmri)
```
