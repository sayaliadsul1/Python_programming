```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

# relational plot

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",style="sex")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker",style="sex",col="day",col_wrap=2,marker=5)
```

# countplot

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.countplot(x="smoker",data=tips)
```

```python
sns.countplot(x ='sex', hue = "smoker", data = tips)
```

```python
sns.countplot(x ='sex', data =tips, palette = "Set2")
```

```python
sns.countplot(y ='sex',hue = "smoker", data = tips)
```

# scatter plot

```python
sns.scatterplot(x='day', y='tip', data=tips,hue="time",marker="+",)
```

```python
sns.scatterplot(x='day', y='tip', data=tips, hue="time", style="time",)
```

# catplot

```python
sns.catplot(x='day', y='total_bill',data=tips,kind='box')
```

# Histogram

```python
sns.histplot(data = tips, x = "size", stat = "probability", discrete = True)
```
