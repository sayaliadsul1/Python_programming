```python
data["MinTemp"].fillna(12.1,inplace=True)
data["MaxTemp"].fillna(23.2,inplace=True)
data["Rainfall"].fillna(2.36,inplace=True)
data["Evaporation"].fillna(5.46,inplace=True)
data["Sunshine"].fillna(7.61,inplace=True)
data["WindGustSpeed"].fillna(40.0,inplace=True)
data["WindSpeed9am"].fillna(14.0,inplace=True)
data["WindSpeed3pm"].fillna(18.6,inplace=True)
data["Humidity9am"].fillna(68.8,inplace=True)
data["Humidity3pm"].fillna(51.5,inplace=True)
data["Pressure9am"].fillna(1017.6,inplace=True)
data["Pressure3pm"].fillna(1015.2,inplace=True)
data["Cloud9am"].fillna(4.44,inplace=True)
data["Cloud3pm"].fillna(4.50,inplace=True)
data["Temp9am"].fillna(16.9,inplace=True)
data["Temp3pm"].fillna(21.6,inplace=True)
```
